# Proper distance in flat single-component universe

```python
# Initial setup
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams.update({'font.size': 14})
rcParams['figure.figsize'] = (10, 5)
```

Plot the current proper distance to a galaxy at redshift $z$ in a flat universe containing one component. Lines are included on the plot for radiation-only ($w = 1/3$), matter-only ($w = 0$), and cosmological constant-only ($w = -1$). Proper distance is in units of $c / H_0$.

```python
# Logarithmically-spaced redshift values from 0.01 (10^-2) to 10000 (10^4)
z = np.logspace(-2, 4, 100)

# Calculate current proper distance
def cpd(z, w):
    dp = 2 / (1 + 3 * w) * (1 - (1 + z)**(-0.5 * (1 + 3 * w)))
    return dp

# Radiation-only universe
wr = 1.0 / 3.0
dpr = cpd(z, wr)
plt.loglog(z, dpr, label='radiation ($w=1/3$)')

# Matter-only universe
wm = 0.0
dpm = cpd(z, wm)
plt.loglog(z, dpm, label='matter ($w=0$)')

# Cosmological constant-only universe
wL = -1.0
dpL = cpd(z, wL)
plt.loglog(z, dpL, label='$\Lambda$ ($w=-1$)')

# Axis labels, legend
plt.xlabel('$z$')
plt.ylabel('$d_p(t_0)$')
plt.legend()
```
